```python
def computeCost(X,y,theta):
    """
   Take the numpy arrays X, y, theta and return the cost function J for this theta. 

    """
    m=len(y)
    h= np.dot(X,theta)
    square_err = (h - y)**2
    J = 1/(2*m) * np.sum(square_err)
    #J = (np.sum(np.square(h - y))) / (2 * m)
    return J
```
